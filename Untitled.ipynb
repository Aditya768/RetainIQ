```python
# ── CHART 3: Churn Risk Score Distribution ────────────────
fig, ax = plt.subplots(figsize=(12, 6))
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

active = df_churn[df_churn['IS_CHURNED'] == False].copy()

# Color bars by risk zone
n, bins, patches = ax.hist(active['CHURN_RISK_SCORE'], bins=40,
                            edgecolor='white', linewidth=0.8)

# Color each bar by risk zone
for patch, left_edge in zip(patches, bins[:-1]):
    if left_edge < 20:
        patch.set_facecolor('#27ae60')
    elif left_edge < 40:
        patch.set_facecolor('#f39c12')
    elif left_edge < 60:
        patch.set_facecolor('#e67e22')
    else:
        patch.set_facecolor('#c0392b')

# Zone labels
ax.axvline(x=20, color='#2c3e50', linestyle='--', alpha=0.6)
ax.axvline(x=40, color='#2c3e50', linestyle='--', alpha=0.6)
ax.axvline(x=60, color='#2c3e50', linestyle='--', alpha=0.6)

ax.text(10,  n.max()*0.92, 'LOW\nRISK',    ha='center', fontsize=10, color='#27ae60', fontweight='bold')
ax.text(30,  n.max()*0.92, 'MEDIUM\nRISK', ha='center', fontsize=10, color='#f39c12', fontweight='bold')
ax.text(50,  n.max()*0.92, 'HIGH\nRISK',   ha='center', fontsize=10, color='#e67e22', fontweight='bold')
ax.text(70,  n.max()*0.92, 'CRITICAL\nRISK', ha='center', fontsize=10, color='#c0392b', fontweight='bold')

# Counts per zone
low      = (active['CHURN_RISK_SCORE'] < 20).sum()
medium   = ((active['CHURN_RISK_SCORE'] >= 20) & (active['CHURN_RISK_SCORE'] < 40)).sum()
high     = ((active['CHURN_RISK_SCORE'] >= 40) & (active['CHURN_RISK_SCORE'] < 60)).sum()
critical = (active['CHURN_RISK_SCORE'] >= 60).sum()

ax.text(10,  n.max()*0.78, f'{low:,} customers',    ha='center', fontsize=9, color='#27ae60')
ax.text(30,  n.max()*0.78, f'{medium:,} customers', ha='center', fontsize=9, color='#f39c12')
ax.text(50,  n.max()*0.78, f'{high:,} customers',   ha='center', fontsize=9, color='#e67e22')
ax.text(70,  n.max()*0.78, f'{critical:,} customers', ha='center', fontsize=9, color='#c0392b')

ax.set_title('Most Customers Are Low Risk — But 167 Critical Risk Accounts Need Immediate Action',
             fontsize=13, fontweight='bold', pad=15)
ax.set_xlabel('Churn Risk Score (0 = Safe, 100 = Certain Churn)', fontsize=12)
ax.set_ylabel('Number of Active Customers', fontsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True, linestyle='--', alpha=0.5)
ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('chart3_churn_risk_score.png', dpi=150, bbox_inches='tight',
            facecolor='white')
plt.show()
print("✅ Chart 3 saved!")
```

```python
# ── CHART 4: Cohort Retention Heatmap ─────────────────────
fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# Reload cohorts
df_cohorts = query("SELECT * FROM RETAINIQ.STAGING.mart_cohorts")

# Pivot for heatmap
cohort_pivot = df_cohorts.groupby(['COHORT_MONTH', 'PLAN'])[
    ['RETENTION_RATE_1M','RETENTION_RATE_3M',
     'RETENTION_RATE_6M','RETENTION_RATE_12M','RETENTION_RATE_18M']
].mean().reset_index()

# Aggregate by cohort month only
cohort_agg = df_cohorts.groupby('COHORT_MONTH')[
    ['RETENTION_RATE_1M','RETENTION_RATE_3M',
     'RETENTION_RATE_6M','RETENTION_RATE_12M','RETENTION_RATE_18M']
].mean().reset_index()

cohort_agg.columns = ['Cohort','1M','3M','6M','12M','18M']
cohort_agg = cohort_agg.sort_values('Cohort').tail(20)
cohort_agg['Cohort'] = cohort_agg['Cohort'].astype(str).str[:7]
cohort_agg = cohort_agg.set_index('Cohort')

# Cast to float
for col in cohort_agg.columns:
    cohort_agg[col] = cohort_agg[col].astype(float)

import seaborn as sns
sns.heatmap(cohort_agg,
            annot=True,
            fmt='.1f',
            cmap='RdYlGn',
            vmin=0, vmax=100,
            linewidths=0.5,
            linecolor='white',
            ax=ax,
            cbar_kws={'label': 'Retention Rate (%)'})

ax.set_title('Retention Drops Sharply After Month 3 — Early Engagement is Critical',
             fontsize=13, fontweight='bold', pad=15)
ax.set_xlabel('Months Since Signup', fontsize=12)
ax.set_ylabel('Cohort (Signup Month)', fontsize=12)
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=9, rotation=0)

plt.tight_layout()
plt.savefig('chart4_cohort_retention.png', dpi=150, bbox_inches='tight',
            facecolor='white')
plt.show()
print("✅ Chart 4 saved!")
```
